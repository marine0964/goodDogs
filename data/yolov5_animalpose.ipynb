```python
name = ['antelope', 'bear', 'bobcat', 'hippocampi', 'orangutan', 'otter', 'rhino']
```

```python
for animal in range(7):
    with open("labels\\" + name[animal] + ".json", "r") as f:
        for st in f.readline().strip('{}]').split('}}], '):
            img_name, bndbox = st.split(': [{')
            img = cv2.imread('images\\' + img_name.strip('"'))
            try:
                img_height, img_width, _ = img.shape
            except:
                print('AttributeError at', img_name)
            
            bnd_list = bndbox.split('}}, {')
            with open("labels\\" + img_name.strip('"').rstrip('jpg') + 'txt', "w") as wr:
                for i in range(len(bnd_list)):
                    temp = bnd_list[i].lstrip('"bndbox": {').split(', ')
                    xmin, xmax, ymax, ymin = map(lambda s: int(s.lstrip('"xymian: ')), temp)
                    
                    #class x_center y_center width height
                    x_center = (xmin + xmax) / 2 / img_width
                    y_center = (ymin + ymax) / 2 / img_height
                    width = min((xmax - xmin) / img_width, 1)
                    height = min((ymax - ymin) / img_height, 1)
                    wr.write(f'{animal} {x_center} {y_center} {width} {height}\n')
```
